Q7. Sum of weights

```python
# Selecting Innjoo laptops
innjoo_laptops = df[df["Brand"] == "Innjoo"]

# Select only the RAM, Storage, Screen columns
innjoo_laptops = innjoo_laptops[['RAM', 'Storage', 'Screen']].copy()

X = innjoo_laptops.values

XTX = X.T.dot(X)

inv_XTX = np.linalg.inv(XTX)

y = np.array([1100, 1300, 800, 900, 1000, 1100])

w = inv_XTX.dot(X.T).dot(y)

sum_w = w.sum()

print(f"The sum of the elements in w: {sum_w}")
```
