```python
import numpy as np
```

```python
# create arrays with zeros, ones and constants

zeros = np.zeros(10)
ones = np.ones(10)
constants = np.full(10, 3.0)
```

```python
print(zeros)
print(ones)
print(constants)
```

```python
# converting lists to arrays

list_example = [1, 2, 3, 4, 5, 6]

list_to_array = np.array(list_example)
```

```python
print(list_to_array)
```

```python
# generating ranges of numbers

np.arange(10)
```

```python
# creating arrays with linear spacing

np.linspace(0, 10, 5)
```

```python
# multi-dimensional arrays

zeros_multi = np.zeros((2, 2))
ones_multi = np.ones((2, 2))
constants_multi = np.full((2, 2), 3)
```

```python
print(zeros_multi)
print(ones_multi)
print(constants_multi)
```

```python
# indexing and slicing arrays

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr)
```

```python
print(arr[0, :])
```

```python
print(arr[:, 0])
```

```python
# generating random arrays

# for reproducibility fix the seed

SEED = 2
np.random.seed(seed=SEED)

# generating random numbers between zero and one
print(np.random.rand(5, 2))

# generate from a normal distribution with mean 0 and std 1
print(np.random.randn(5, 2))

# generate from a random distribution of integers
print(np.random.randint(low = 0, high = 100, size = (5, 2)))
```

```python
# Array operations

# element wise operations on one array
print(arr)
arr = arr + 1
print(arr)
arr = arr * 2
print(arr)
arr = arr / 2
print(arr)
arr = arr - 1
print(arr)
```

```python
# element wise operations between multiple arrays

arr1 = np.ones(4)
arr2 = np.full(4, 3)

print(arr1 / arr2)
print(arr1 * arr2)
print(arr1 + arr2)
print(arr1 - arr2)
```

```python
# comparison operations, filtering operations

arr = np.array([1, 2, 3, 4])
greater_than_2 = arr > 2

print(arr)
print(greater_than_2)
```

```python
selected_elements = arr[arr > 1]  # Gets elements greater than 1
```

```python
selected_elements
```

```python
arr
```

```python
# summary operations

min_value = arr.min()    # Minimum value
print(min_value)
max_value = arr.max()    # Maximum value
print(max_value)
sum_value = arr.sum()    # Sum of all elements
print(sum_value)
mean_value = arr.mean()  # Mean (average) value
print(mean_value)
std_deviation = arr.std()  # Standard deviation
print(std_deviation)
```
